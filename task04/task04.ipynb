## PAI Lab 4 Task 1 N-Queens Problem (dynamic N) using backtracking

## Maham Nafees 
## Roll Number: 038

```python
n = 8

def ok(pos, r, c):
    for rr, cc in enumerate(pos):
        if cc == c:
            return False
        if abs(rr - r) == abs(cc - c):
            return False
    return True

def solve(r, pos):
    if r == n:
        return pos
    for c in range(n):
        if ok(pos, r, c):
            res = solve(r + 1, pos + [c])
            if res is not None:
                return res
    return None

sol = solve(0, [])
if sol is None:
    print("no solution")
else:
    for r in range(n):
        row = []
        for c in range(n):
            row.append("Q" if sol[r] == c else ".")
        print(" ".join(row))
```
