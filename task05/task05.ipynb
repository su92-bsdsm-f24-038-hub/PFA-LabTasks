```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img_path = "university.jpg"

img = cv2.imread(img_path)
if img is None:
    print("image not found, set img_path to a real image")
else:
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # get image size
    h, w = rgb.shape[:2]
    small = cv2.resize(rgb, (w//2, h//2))
    # crop the top-left quarter of the image
    crop = rgb[0:h//2, 0:w//2]
    #gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    _, th = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

    edges = cv2.Canny(gray, 80, 160)

    draw = rgb.copy()
    cv2.rectangle(draw, (10, 10), (120, 80), (255, 0, 0), 2)
    cv2.circle(draw, (180, 120), 35, (0, 255, 0), 2)
    cv2.line(draw, (0, 0), (w-1, h-1), (0, 0, 255), 2)
    cv2.putText(draw, "demo", (10, h-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,0), 2)

    figs = [("rgb", rgb), ("small", small), ("crop", crop), ("gray", gray), ("th", th), ("edges", edges), ("draw", draw)]
    for title, im in figs:
        plt.figure()
        if len(im.shape) == 2:
            plt.imshow(im, cmap="gray")
        else:
            plt.imshow(im)
        plt.title(title)
        plt.axis("off")
    plt.show()
```
