### Task 1 - Tokenization

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize, sent_tokenize

text = "Natural language processing is fun. I am learning NLP today."

words = word_tokenize(text)
sentences = sent_tokenize(text)

print("Word Tokens:", words)
print("Sentence Tokens:", sentences)
```

### Task 2 - Stopwords Removal

```python
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

sentence = "This is a simple example to show stopwords removal"
words = word_tokenize(sentence)

filtered = []
for w in words:
    if w not in stop_words:
        filtered.append(w)

# printing the filered words
print("Original:", words)
print("After Removing Stopwords:", filtered)
```

### Task 3 - Stemming

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "played", "flies", "happily", "studies"]

for word in words:
    print(word, "-->", stemmer.stem(word))
```

### Task 4 - Lemmatization

```python
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["running", "better", "flies", "studies", "played"]

# lemmatizing each word in the list
for word in words:
    print(word, "-->", lemmatizer.lemmatize(word, pos='v'))
```

### Task 5 - Sentiment Analysis using TextBlob

```python
from textblob import TextBlob

reviews = [
    "This product is excellent!",
    "I hate this so much.",
    "It is okay, not bad not good."
]

for review in reviews:
    blob = TextBlob(review)
    polarity = blob.sentiment.polarity

    if polarity > 0:
        result = "Positive"
    elif polarity < 0:
        result = "Negative"
    else:
        result = "Neutral"

    # printing polarity and sentiment
    print(f"Review: {review}")
    print(f"Polarity: {polarity} --> {result}")
    print()
```

### Task 6 - Text Classification using Sklearn

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

train_data = [
    "I love this movie",
    "This film is amazing",
    "I hate this movie",
    "This was a terrible film",
    "Absolutely wonderful experience",
    "Worst movie ever"
]

lables = ["positive", "positive", "negative", "negative", "positive", "negative"]

vectorizer = CountVectorizer()
model = MultinomialNB()

x_train = vectorizer.fit_transform(train_data)
model.fit(x_train, lables)

test_data = ["I hate it", "This was great", "Not good at all"]

x_test = vectorizer.transform(test_data)
predictions = model.predict(x_test)

for i in range(len(test_data)):
    print(f"Text: {test_data[i]} --> {predictions[i]}")
```
