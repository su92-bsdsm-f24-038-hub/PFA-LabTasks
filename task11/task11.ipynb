### 1. LangChain

LangChain is a framework that helps developers build applications using Large Language Models (LLMs). It provides tools and components to connect LLMs with external data sources, APIs, memory, and other tools.

Instead of just sending a single prompt to an LLM, LangChain allows you to create **chains** of operations. For example, you can retrieve data from a database, pass it to an LLM, and then format the output — all in one pipeline.

**Key Features:**
- Chains: Sequence of calls to LLMs or other tools
- Agents: LLMs that can decide which tools to use
- Memory: Stores conversation history
- Integrations: Works with OpenAI, HuggingFace, FAISS, ChromaDB, etc.

**Example Use Case:** Building a chatbot that reads from a PDF and answers questions about it.

```python
from langchain.llms import OpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

llm = OpenAI(temperature=0.7)
prompt = PromptTemplate(input_variables=["topic"], template="Explain {topic} in simple words.")
chain = LLMChain(llm=llm, prompt=prompt)

print(chain.run("machine learning"))
```

### 3. RAG (Retrieval-Augmented Generation)

RAG stands for **Retrieval-Augmented Generation**. It is a technique that combines the power of LLMs with a search/retrieval system. Instead of relying only on what the model learned during training, RAG allows the model to **fetch relevant information** from an external source (like a document or database) before generating a response.

**How RAG Works:**
1. User asks a question
2. The system searches a document store for relevant chunks
3. The retrieved chunks are passed to the LLM as context
4. The LLM generates an answer based on the retrieved context

**Why RAG is useful:**
- LLMs have a knowledge cutoff date, RAG solves this by fetching live data
- Reduces hallucinations (wrong answers made up by the model)
- Can work with private/company documents

**Example Use Case:** A chatbot that answers questions from a company's internal PDF documents.

FAISS is very fast even with millions of vectors, which makes it ideal for production AI systems.

### 5. Vector

In AI, a **vector** is a list of numbers that represents some data (like a word, sentence, or image) in a mathematical form. These numbers capture the **meaning or features** of the data.

For example, the word "apple" might be represented as:
```
"apple" --> [0.21, -0.34, 0.76, 0.05, ..., 0.02]
```

Words or sentences with similar meanings will have vectors that are **close to each other** in vector space. This is called a **semantic similarity**.

**How vectors are created:**
- Word2Vec, GloVe — for word embeddings
- BERT, Sentence-BERT — for sentence embeddings
- CLIP — for image embeddings

**Example using Word2Vec:**

```python
from gensim.models import Word2Vec

sentences = [["king", "queen", "man", "woman"], ["paris", "france", "berlin", "germany"]]
model = Word2Vec(sentences, min_count=1)

# getting the vector for king
vector = model.wv["king"]
print(vector)
```

Vectors are the foundation of modern AI — almost everything in deep learning is represented as vectors.

### 6. VectorDB (Vector Database)

A **VectorDB** is a special type of database designed to store, index, and search **vectors (embeddings)** efficiently. Unlike traditional databases that search by exact match, VectorDBs search by **similarity** — finding vectors that are closest to a given query vector.

**Why VectorDB is needed:**
- Normal databases like MySQL cannot do similarity search on vectors
- VectorDBs are optimized for high-dimensional vector search
- They are the backbone of RAG systems and semantic search

**Popular VectorDBs:**
| Database | Description |
|----------|-------------|
| ChromaDB | Open-source, easy to use locally |
| Pinecone | Cloud-based, production ready |
| Weaviate | Open-source with GraphQL support |
| Qdrant | Fast and lightweight |
| FAISS | Library (not a full DB but used similarly) |

**Example with ChromaDB:**

```python
import chromadb

client = chromadb.Client()
collection = client.create_collection("my_docs")

collection.add(
    documents=["AI is the future", "Python is a great language"],
    ids=["doc1", "doc2"]
)

results = collection.query(query_texts=["What is AI?"], n_results=1)
print(results)
```

### 7. Generative AI (GenAI)

**Generative AI** refers to AI systems that can **create new content** — such as text, images, audio, video, or code — by learning patterns from existing data.

Unlike traditional AI which only classifies or predicts, GenAI actually **generates** new and original outputs.

**Types of Generative AI:**
- **Text Generation:** ChatGPT, GPT-4, LLaMA
- **Image Generation:** DALL·E, Stable Diffusion, Midjourney
- **Code Generation:** GitHub Copilot, Codex
- **Audio/Music:** MusicLM, AudioCraft
- **Video:** Sora (OpenAI)

**How GenAI works (simplified):**
1. Model is trained on huge amounts of data
2. It learns the patterns and structure of that data
3. When given a prompt, it generates new content that follows those patterns

**Example using HuggingFace pipeline:**

```python
from transformers import pipeline

# text generation using GPT-2
gen = pipeline("text-generation", model="gpt2")
result = gen("Once upon a time in a land far away", max_length=60)
print(result[0]["generated_text"])
```

GenAI is one of the fastest growing areas in AI and is changing how we create and interact with digital content.

### 8. GANs (Generative Adversarial Networks)

**GANs** are a type of generative AI model introduced by Ian Goodfellow in 2014. They consist of **two neural networks** that compete against each other:

- **Generator:** Tries to create fake data that looks real
- **Discriminator:** Tries to tell the difference between real and fake data

These two networks train together. The generator keeps improving to fool the discriminator, and the discriminator keeps improving to catch the generator. Over time, the generator becomes very good at creating realistic data.

**How GANs work:**
```
Random Noise --> Generator --> Fake Image
                                  
```

**Applications of GANs:**
- Generating realistic human faces (e.g., thispersondoesnotexist.com)
- Image-to-image translation (e.g., sketch to photo)
- Data augmentation for training datasets
- Video generation and deepfakes
- Art and creative content generation

**Simple GAN structure in Keras:**

```python
from keras.models import Sequential
from keras.layers import Dense

# Generator model
generator = Sequential()
generator.add(Dense(128, input_dim=100, activation='relu'))
generator.add(Dense(784, activation='sigmoid'))

# Discriminator model
discriminator = Sequential()
discriminator.add(Dense(128, input_dim=784, activation='relu'))
discriminator.add(Dense(1, activation='sigmoid'))

print("Generator Summary:")
generator.summary()

print("Discriminator Summary:")
discriminator.summary()
```

GANs are powerful but can be difficult to train because both networks need to improve at the same rate. If one becomes too strong, the training becomes unstable.
