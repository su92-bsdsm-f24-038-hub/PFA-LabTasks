## PAI Lab 3 Task 1 Water Jug Problem using DFS (4-liter jug and 3-liter jug, goal 2 liters)

```python
cap1 = 4
cap2 = 3
goal = 2

def nxt(st):
    a, b = st
    out = []
    out.append(((cap1, b), "fill jug1"))
    out.append(((a, cap2), "fill jug2"))
    out.append(((0, b), "empty jug1"))
    out.append(((a, 0), "empty jug2"))

    p = min(a, cap2 - b)
    out.append(((a - p, b + p), "pour jug1 to jug2"))

    p = min(b, cap1 - a)
    out.append(((a + p, b - p), "pour jug2 to jug1"))
    return out

def dfs():
    st0 = (0, 0)
    stack = [(st0, [])]
    seen = set()

    while stack:
        st, path = stack.pop()
        if st in seen:
            continue
        seen.add(st)

        a, b = st
        if a == goal or b == goal:
            return path + [(st, "goal")]

        for ns, act in nxt(st):
            if ns not in seen:
                stack.append((ns, path + [(st, act)]))
    return None

ans = dfs()
if ans is None:
    print("no solution found")
else:
    for st, act in ans:
        print(st, "->", act)
```
